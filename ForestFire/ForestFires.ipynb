```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
forest_fire = pd.read_csv("ForestFires.csv") 
forest_fire.head()
```

```python
plt.figure(figsize = (11.7,8.27))
pv_table = pd.pivot_table(forest_fire,
                       values = 'X', 
                       index = 'month',
                       aggfunc='count',
                       fill_value = 0)

order_month = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
pv_table = pv_table.reindex(order_month)
plt.bar(pv_table.index, pv_table['X'])
plt.xlabel('Months')
plt.ylabel('Count of Forest Fires')
plt.title('Number of Forest Fires')
plt.show()
```

```python
ax = sns.pairplot(forest_fire[['temp','RH','wind','rain','area']], diag_kind="kde")
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
corr_matrix = forest_fire[['temp','RH','wind','rain','area']].corr()
ax = sns.heatmap(corr_matrix, annot = True)
```

```python
plot=sns.scatterplot(data=forest_fire, x= 'wind', y= 'area', color= 'steelblue')
plot.set(ylim=(0, 300))
plt.title("Relationship between Wind Speed and Area of Fire")
plt.show()
```
